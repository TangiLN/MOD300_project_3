# Description of the testing Method 

### 
To calculate the volume of DNA with the  random walk method, We choose to set a number of starting point for each walk. We make the walker go in a random direction in the simulation box and when it enter in collision with a sphere or the edge of the box we stop the walker. 
We store the path of each walker to estimate the volume outisde the sphere and that way we can calculate the volume without sphere on the total volume and have the volume of DNA. 
A test could be to use this method on specifique sphere or specific cases,  and  test it to see if we have the expected result. And also to calculate the analytical volume.
