## Description of the testing Method 

###
To calculate the volume of DNA with the  random walk method, We chose to set a number of starting point for each walk. We make the walker go in a random direction in the simulation box and when it enters in collision with a sphere or the edge of the box we stop the walker. 
We store the path of each walker to estimate the volume outisde the sphere and that way we can calculate the volume without sphere on the total volume and have the volume of DNA.  

To do that at first we have to add a function that generates random starting point for the walkers and after that, a function that makes them move and store the data of the spheric and non spheric values.
